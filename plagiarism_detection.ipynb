```python
import spacy
from transformers import BertTokenizer, BertForSequenceClassification
import torch
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load the English tokenizer and model for dependency parsing
nlp = spacy.load("en_core_web_sm")

```

```python
# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')

# Set the model to evaluation mode
model.eval()

def dependency_parse(text):
    doc = nlp(text)
    dependencies = [(token.text, token.dep_) for token in doc]
    return dependencies
```

```python
def classify_plagiarism(original_text, suspicious_text):
    # Extract dependency parses
    original_dependencies = dependency_parse(original_text)
    suspicious_dependencies = dependency_parse(suspicious_text)

    # Compare dependency parses (simplified example)
    similarity_score = 0
    for original_token, suspicious_token in zip(original_dependencies, suspicious_dependencies):
        if original_token == suspicious_token:
            similarity_score += 1

    # Convert texts to BERT input format
    inputs = tokenizer(original_text, suspicious_text, return_tensors="pt", padding=True, truncation=True)

    # Perform plagiarism classification using BERT
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.argmax(outputs.logits, dim=1)

    # Return plagiarism classification result and similarity score
    return predictions.item(), similarity_score / len(original_dependencies)

```

```python
# Function to read text files from a folder
def read_files_from_folder(folder_path):
    texts = []
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.txt'):
            file_path = os.path.join(folder_path, file_name)
            with open(file_path, 'r') as file:
                text = file.read()
                texts.append(text)
    return texts

```

```python
# Example usage comparing two folders
original_folder_path = '/content/drive/MyDrive/datasets/dataset/dataset for plag/dataset/Dokumen Teks/Original'
suspicious_folder_path = '/content/drive/MyDrive/datasets/dataset/dataset for plag/dataset/Dokumen Teks/Copy'

# Read text files from each folder
original_texts = read_files_from_folder(original_folder_path)
suspicious_texts = read_files_from_folder(suspicious_folder_path)
```
